```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

x = 2 * np.random.rand(100, 1)

y = 4 + 3 * x + np.random.randn(100, 1)
```

```python
plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Generated Data")
plt.show()
```

```python
X_b = np.c_[np.ones((100, 1)), x]

theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

print(f"Intercept: {theta_best[0][0]}, Slope: {theta_best[1][0]}")
```

```python
# Create new data points for prediction
x_new = np.array([[0], [2]])

# Add a column of ones to x_new
X_new_b = np.c_[np.ones((2, 1)), x_new]

# Predict y values using the model
y_predict = X_new_b.dot(theta_best)

# Plot the predictions along with the original data
plt.plot(x_new, y_predict, "r-", label="Predictions")
plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression Fit")
plt.legend()
plt.show()
```

```python
# Predict y for the original x values
y_pred = X_b.dot(theta_best)

# Calculate Mean Squared Error (MSE)
mse = np.mean((y - y_pred) ** 2)
print(f"Mean Squared Error: {mse}")
```
